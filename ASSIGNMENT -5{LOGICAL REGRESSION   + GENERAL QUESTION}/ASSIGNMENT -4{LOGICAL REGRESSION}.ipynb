```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import sklearn.metrics
from sklearn.datasets import load_breast_cancer
from sklearn.datasets import load_iris
```

```python
## TASK 5 : : 
```

```python
# A 
bc_data = load_breast_cancer()
features = bc_data.data
targets = bc_data.target
```

```python
#b
xtrain, xtest, ytrain, ytest = train_test_split(features, targets, test_size=0.2, random_state=42)

# Build and train model
lr_clf = LogisticRegression(max_iter=10000)
lr_clf.fit(xtrain, ytrain)

# Predict labels and probabilities
pred_class = lr_clf.predict(xtest)
pred_prob = lr_clf.predict_proba(xtest)[:,1]
```

```python
#c
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

acc_score = accuracy_score(ytest, pred_class)
cm = confusion_matrix(ytest, pred_class)
report = classification_report(ytest, pred_class)

# Calculate ROC-AUC score correctly
roc_auc = roc_auc_score(ytest, pred_prob)

print("Accuracy:", acc_score)
print("Confusion Matrix:")
print(cm)
print("Classification Report:")
print(report)
print('\nROC-AUC Score:', roc_auc)

# Compute ROC curve points
fpr_vals, tpr_vals, _ = roc_curve(ytest, pred_prob)

plt.plot(fpr_vals, tpr_vals, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], '--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC CURVE')
plt.legend()
plt.grid()
plt.show()
```

```python
#a) Predict the Probability of the positive class
y_prob = lr_clf.predict_proba(xtest)[:, 1]
```

```python
# Show effect of changing probability threshold
from sklearn.metrics import f1_score, confusion_matrix

def check_threshold(thresh):
    pred_thresh = (pred_prob >= thresh).astype(int)
    matrix = confusion_matrix(ytest, pred_thresh)
    score = f1_score(ytest, pred_thresh)
    print(f"Threshold: {thresh}")
    print("Confusion Matrix:\n", matrix)
    print("F1 Score:", round(score, 3))
    print()

for value in [0.3, 0.5, 0.7]:
    check_threshold(value)
```

```python
# Compute optimal threshold using Youden's J statistic
fpr2, tpr2, threshs = roc_curve(ytest, pred_prob)
j_stat = tpr2 - fpr2
j_best = np.argmax(j_stat)
opt_thresh = threshs[j_best]

plt.figure(figsize=(7,5))
plt.plot(fpr2, tpr2, label='ROC Curve')
plt.scatter(fpr2[j_best], tpr2[j_best], color='red', label='Best Threshold')
plt.plot([0,1], [0,1], '--', color='grey')
plt.title("ROC Curve with opt Threshold")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid()
plt.show()
print("Optimal Threshold (Youden's J):", round(opt_thresh, 3))
```

```python
# Load Iris data
iris = load_iris()
x_iris = iris.data
y_iris = iris.target
names = iris.target_names
```

```python

# Split and train
xtr_iris, xte_iris, ytr_iris, yte_iris = train_test_split(x_iris, y_iris, test_size=0.25, random_state=42)
iris_lr = LogisticRegression(max_iter=1000)
iris_lr.fit(xtr_iris, ytr_iris)
iris_pred = iris_lr.predict(xte_iris)
```

```python
# Evaluate
print("Accuracy (Iris):", accuracy_score(yte_iris, iris_pred))
print("Classification Report (Iris):\n", classification_report(yte_iris, iris_pred, target_names=names))
```
